```python
import matplotlib.pyplot as plt
import numpy as np 
from scipy.stats import beta
```

```python
plt.style.use('ggplot')
```

## Beta Shape

```python
beta.pdf([0, 1], 1, 2)
```

```python
alpha_list = [1, 3, 20, 50]
beta_list = [2, 3, 10, 20]
x = np.arange(0, 1.001, .001)
y_list = []
for a, b in zip(alpha_list, beta_list):
    y_list.append(beta.pdf(x, a, b))
```

```python
for i, a, b in zip(range(4), alpha_list, beta_list):
    plt.plot(x, y_list[i], label=f"alpha = {a}, beta = {b}")
plt.legend()
plt.show()
```
